```python
import pandas as pd 
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv' )
```

```python
data
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(x = data['Median'], y = data['Unemployment_rate'])
plt.xlabel('Median')
plt.ylabel('Unemployment_rate')
plt.show()
```

#### As seen easily from the scatter chart, a corelation does not exist between the earnings and unemployment.


```python
data[data["Median"] > 60000]
```

```python
median=data[data["Median"] > 60000].sort_values('Median')
median.plot(x='Major',y=['P25th','Median','P75th'],kind='bar')
```

#### Yes, it is true that petroleum engineering majors were by far the best paid recent graduates as clearly shown in the bar chart.
